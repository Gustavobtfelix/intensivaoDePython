- Nosso desafio é conseguir prever as vendas que vamos ter em determinado período com base nos gastos em anúncios nas 3 grandes redes que a empresa Hashtag investe: TV, Jornal e Rádio
- TV, Jornal e Rádio estão em milhares de reais
- Vendas estão em milhões

#### Importar a Base de dados

```python
import pandas as pd

tabela = pd.read_csv("advertising.csv")
display(tabela)
tabela.info()
```

#### Com isso, podemos partir para a preparação dos dados para treinarmos o Modelo de Machine Learning

- Separando em dados de treino e dados de teste

```python
#Separar dados em base de treino e base de teste

y = tabela["Vendas"] #quem vai prever
x = tabela[["TV", "Radio", "Jornal"]] #Referencia para a previsão
#ou x = tabela.drop("Vendas", axis=1)

#separar dados em treino e teste sklearn é uma biblioteca de machine learning
from sklearn.model_selection import train_test_split
#train_test_split() separa 80% da base para treino e 20% para teste. test_size=0.3 altera a alocação da base para teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)

```

#### Temos um problema de regressão - Vamos escolher os modelos que vamos usar:

- Regressão Linear
- RandomForest (Árvore de Decisão)

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

#Criou 2 inteligencia artificiais para comparar e ver qual faz a melhor previsão
modeloRegressaoLinear = LinearRegression()
modeloArvoreDecisao = RandomForestRegressor()

#Treinou as inteligencias artificiais
modeloRegressaoLinear.fit(x_treino, y_treino)
modeloArvoreDecisao.fit(x_treino, y_treino)
```

#### Teste da AI e Avaliação do Melhor Modelo

- Vamos usar o R² -> diz o % que o nosso modelo consegue explicar o que acontece

```python
#Etapa de testes

regressaoLinear = modeloRegressaoLinear.predict(x_teste)
arvoreDecisao = modeloArvoreDecisao.predict(x_teste)

from sklearn import metrics

print(metrics.r2_score(y_teste, regressaoLinear)) #compara o valor do y_teste com a previsão da inteligencia artificial
print(metrics.r2_score(y_teste, arvoreDecisao))

# a arvore de decisão é o grafico mais preciso
```

#### Visualização Gráfica das Previsões

```python
tabela_auxiliar = pd.DataFrame()#cria uma tabela vazia
tabela_auxiliar['y_teste'] = y_teste #cria a coluna y_teste e adiciona os valores de vendas que foram separados do bd no train_test_split
tabela_auxiliar['Previsao Regressao Linear'] = regressaoLinear
tabela_auxiliar['Previsao Arvore Decisao'] = arvoreDecisao

#Tamanho da Figura
plt.figure(figsize=(15,6))
#seaborn
sns.lineplot(data=tabela_auxiliar)
plt.show()
```
